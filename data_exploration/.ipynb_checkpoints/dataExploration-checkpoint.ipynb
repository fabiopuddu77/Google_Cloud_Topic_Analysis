```python
#LEGGO I FILE PARQUET  - 1.1
from google.cloud import storage
import pandas as pd
from io import BytesIO

# Initialize Google Cloud Storage client
client = storage.Client()

# Your GCS bucket name
bucket_name = 'ccai-storage'

# Define the path within the bucket
blob_name = 'pipeline/articlesoutput.parquet'  # Adjust this to your file's GCS path

# Create a bucket object and a blob (file) object
bucket = client.bucket(bucket_name)
blob = bucket.blob(blob_name)

# Download the blob contents into memory
blob_contents = blob.download_as_bytes()

# Use BytesIO to open the blob contents as a file-like object, which pandas can read from
articles_df = pd.read_parquet(BytesIO(blob_contents))

# Now `dataframe` holds your data, and you can work with it
```

1.3 ANALISI CORRELAZIONI E ASSOCIAZIONI CON VARIABILE TARGET SHARESFACEBOOK

```python
#1.3 ESAME CATEGORIES: ANOVA TEST

import scipy.stats as stats
import pandas as pd

# Count the number of articles for each category (considering combined categories as distinct)
category_counts = df_2023['topic'].value_counts()

# Count the total number of articles
total_articles = len(df_2023)

print("Number of articles for each category:")
print(category_counts)
print("\nTotal number of articles:", total_articles)

# Assuming articles_df is your DataFrame containing the articles data
# First, filter out articles with null or empty 'categoryLabels'
filtered_articles_df = df_2023[df_2023['topic'].notna() & (df_2023['topic'] != '')]

# Then, perform the ANOVA test
anova_result = stats.f_oneway(*[group['sharesFacebook'].dropna() for name, group in filtered_articles_df.groupby('topic')])

print(f"ANOVA result: F-statistic = {anova_result.statistic}, p-value = {anova_result.pvalue}")
```

```python
#1.3 TEST KRUSKAL-wallis tra category e label 
import scipy.stats as stats

# Assuming filtered_articles_df is your DataFrame that has been filtered to exclude empty categories
# And assuming 'categoryLabels' does not need to be exploded (as per your request)

# Perform the Kruskal-Wallis H-test on 'sharesFacebook' across the groups defined by 'categoryLabels'
kruskal_result = stats.kruskal(*[group['sharesFacebook'].dropna() for name, group in filtered_articles_df.groupby('topic')])

print(f"Kruskal-Wallis H-test result: H-statistic = {kruskal_result.statistic}, p-value = {kruskal_result.pvalue}")

# Calculate median sharesFacebook and the count of articles for each category
category_stats = filtered_articles_df.groupby('topic')['sharesFacebook'].agg(['median', 'count'])

# Sort the categories by median sharesFacebook for better visualization
sorted_category_stats = category_stats.sort_values(by='median', ascending=False)

# Print the top N categories by median sharesFacebook, along with their counts
top_n = 30  # Adjust N based on how many you want to see
print(f"Top {top_n} Categories by Median sharesFacebook and Article Count:")
print(sorted_category_stats.head(top_n))
```

```python
import pandas as pd
from scipy import stats

# Assuming your DataFrame is already prepared and includes 'is_weekend' and 'shares_facebook'
# df = pd.read_parquet('df.parquet')  # Load your DataFrame here

# Separate shares_facebook into two groups: weekend vs. other days
shares_weekend = df_2023[df_2023['is_weekend']]['sharesFacebook']
shares_other_days = df_2023[~df_2023['is_weekend']]['sharesFacebook']

# Perform a t-test
t_stat, p_value = stats.ttest_ind(shares_weekend, shares_other_days, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("There is a significant difference in shares_facebook between weekends and other days.")
else:
    print("There is no significant difference in shares_facebook between weekends and other days.")
```
